## Introducción al problema

El examen Saber 11, dirigido por el ICFES, es la prueba que marca la culminación de la etapa académica de los jóvenes en Colombia y evalúa competencias en Matemáticas, Lectura Crítica, Sociales y Ciudadanas, Ciencias Naturales e Inglés. Además de los resultados, la prueba recoge información económica, académica, social y demográfica de los estudiantes, lo que permite analizar cómo factores como pertenecer a una minoría étnica, el acceso a las TIC o la zona de residencia influyen en el desempeño. Este enfoque integral facilita la identificación de poblaciones en situación de vulnerabilidad y brinda insumos para orientar políticas de apoyo, como becas o subsidios, con el fin de reducir desigualdades educativas en el país.

## Pregunta SMART

¿En qué medida las características sociodemográficas, económicas y académicas de los evaluados en el examen Saber 11 del año 2024 se relacionan con su puntaje global obtenido en la prueba, y cómo podrían estos hallazgos orientar estrategias de apoyo educativo focalizado?

## Encoding y manejo de variables

En esta sección buscamos dejar listas nuestras variables en formatos útiles para el ajuste del modelo ML.

Primero revisamos si las columnas de nuestro dataset si presentan los valores definidos en el diccionario de datos y dar un vistazo de cómo se comportan para darnos una idea de que tipo de encoding aplicar.

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"\nUnique values for '{col}':")
    display(df[col].value_counts(dropna=False))
```

Al revisar los datos, confirmamos que ninguna variable presenta valores fuera de los establecidos en el diccionario. Además, observamos que muchas de ellas son de carácter binario o poseen un orden natural entre sus categorías.

Sin embargo, identificamos una excepción en las variables fami_educacionpadre, fami_educacionmadre, fami_trabajolaborpadre y fami_trabajolabormadre, ya que contienen valores como “No aplica” y “No sabe”, los cuales no se ajustan a la escala ordinal definida en el resto de categorías. Por esta razón, tomamos la decisión de descartar las filas que presentan dichos valores.

```python
df = df[~df["fami_educacionmadre"].isin(["No Aplica", "No sabe"])]
df = df[~df["fami_educacionpadre"].isin(["No Aplica", "No sabe"])]
df = df[~df["fami_trabajolaborpadre"].isin(["No aplica", "No sabe"])]
df = df[~df["fami_trabajolabormadre"].isin(["No aplica", "No sabe"])]
```

Se decide abordar la predicción del puntaje como una situación de clasificación, debido a que al ajustar modelos de regresión lineal se obtuvieron métricas como el R2 alrededor del 34%. Es por esto que modificamos la variable puntaje global tal que tuviera los niveles Bajo, Medio y Alto, dividiendo los datos en 3 terciles o grupos donde en cada uno se encuentra alrededor del 33.3% de los datos.

```python
# Crear grupos de puntaje
df['grupo_puntaje'] = pd.qcut(df['punt_global'], q=3, labels=['Bajo','Medio','Alto'])

# Promedio de variables por grupo
promedios = df.groupby('grupo_puntaje').mean(numeric_only=True).T
print(promedios.head())
```

Se elige modelar el problema de clasificación mediante el ajuste de un Random Forest donde se exploran diferentes hiperparámetros  utilizando validación cruzada para hallar el mejor modelo.

```python

# ==========================================
# 1. Definición del espacio de búsqueda
# ==========================================
param_dist = {
    'n_estimators': [200, 300, 500, 800],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_features': ['sqrt', 'log2']
}

# Modelo base
rf = RandomForestClassifier(random_state=42, n_jobs=-1)


# ==========================================
# 2. Preparación de datos
# ==========================================
# Convertimos booleans -> int
X = df.drop(columns=['punt_global', 'grupo_puntaje'])
X = X.assign(**{c: X[c].astype(int) for c in X.select_dtypes('bool')})

y = df['grupo_puntaje']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)


# ==========================================
# 3. RandomizedSearch con CV
# ==========================================
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=30,               # número de combinaciones a probar
    cv=5,                    # validación cruzada 5-fold
    scoring='accuracy',      # se puede cambiar a f1_weighted, balanced_accuracy
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

# ==========================================
# 4. Resultados
# ==========================================
print("Mejores parámetros:", random_search.best_params_)
print("Mejor exactitud CV:", random_search.best_score_)

best_model = random_search.best_estimator_

# Exactitud en hold-out final (train/test aparte si se quiere comparar)
#X_train, X_test, y_train, y_test = train_test_split(
#    X, y, test_size=0.3, stratify=y, random_state=42
#)
print("Exactitud en test con mejor modelo:", best_model.score(X_test, y_test))
```

Ahora se ajusta el modelo con los hiperparámetros hallados que conducen al mejor modelo.

```python
# ==========================================
# 1. Preparación de datos
# ==========================================
# Convertimos booleans -> int
X = df.drop(columns=['punt_global', 'grupo_puntaje'])
X = X.assign(**{c: X[c].astype(int) for c in X.select_dtypes('bool')})

y = df['grupo_puntaje']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ==========================================
# 2. Entrenamiento Random Forest con mejores parámetros
# ==========================================
rf = RandomForestClassifier(
    n_estimators=800,
    max_depth=20,
    min_samples_split=2,
    min_samples_leaf=5,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

# Score
print("Exactitud en train:", rf.score(X_train, y_train))
print("Exactitud en test :", rf.score(X_test, y_test))

# ==========================================
# 3. Importancia de variables
# ==========================================
importancias = pd.DataFrame({
    'Variable': X.columns,
    'Importancia': rf.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print("\nTop 10 variables más importantes:")
print(importancias.head(10))

# ==========================================
# 4. Gráfico de importancia
# ==========================================
plt.figure(figsize=(10,6))
sns.barplot(
    data=importancias.head(15),
    x='Importancia', y='Variable',
    palette="viridis"
)
plt.title("Top 15 factores que influyen en el puntaje (Random Forest)")
plt.xlabel("Importancia")
plt.ylabel("Variable")
plt.tight_layout()
plt.show()
```
