## Introducción al problema

El examen Saber 11, dirigido por el ICFES, es la prueba que marca la culminación de la etapa académica de los jóvenes en Colombia y evalúa competencias en Matemáticas, Lectura Crítica, Sociales y Ciudadanas, Ciencias Naturales e Inglés. Además de los resultados, la prueba recoge información económica, académica, social y demográfica de los estudiantes, lo que permite analizar cómo factores como pertenecer a una minoría étnica, el acceso a las TIC o la zona de residencia influyen en el desempeño. Este enfoque integral facilita la identificación de poblaciones en situación de vulnerabilidad y brinda insumos para orientar políticas de apoyo, como becas o subsidios, con el fin de reducir desigualdades educativas en el país.

## Pregunta SMART

¿En qué medida las características sociodemográficas, económicas y académicas de los evaluados en el examen Saber 11 del año 2024 se relacionan con su puntaje global obtenido en la prueba, y cómo podrían estos hallazgos orientar estrategias de apoyo educativo focalizado?

Analizando **estu_tieneetnia** se puede apreciar que solo presenta valores de S y Si y lo demás es nulo. Con esto y revisando el diccionario de datos podríamos asumir que estos valores nulos representan una respuesta negativa "No" a la pregunta de si el estudiante pertenece a un grupo etnico.

```python
print("\nConteo de valores únicos para 'estu_etnia':")
display(df['estu_etnia'].value_counts(dropna=False))
```

Una situación parecida se repite en **estu_etnia**, se puede apreciar como los valores no nulos hacen referencia a distintos grupos étnicos alrededor del país y otra vez podríamos hacer la asumpción de que un valor nulo en esta variable representa que el estudiante no hace parte de ningún grupo etnico.

## Manejo de columnas

## Encoding y manejo de variables

En esta sección buscamos dejar listas nuestras variables en formatos útiles para el ajuste del modelo ML.

Primero revisamos si las columnas de nuestro dataset si presentan los valores definidos en el diccionario de datos y dar un vistazo de cómo se comportan para darnos una idea de que tipo de encoding aplicar.

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"\nUnique values for '{col}':")
    display(df[col].value_counts(dropna=False))
```

Al revisar los datos, confirmamos que ninguna variable presenta valores fuera de los establecidos en el diccionario. Además, observamos que muchas de ellas son de carácter binario o poseen un orden natural entre sus categorías.

Sin embargo, identificamos una excepción en las variables fami_educacionpadre, fami_educacionmadre, fami_trabajolaborpadre y fami_trabajolabormadre, ya que contienen valores como “No aplica” y “No sabe”, los cuales no se ajustan a la escala ordinal definida en el resto de categorías. Por esta razón, tomamos la decisión de descartar las filas que presentan dichos valores.

```python
df = df[~df["fami_educacionmadre"].isin(["No Aplica", "No sabe"])]
df = df[~df["fami_educacionpadre"].isin(["No Aplica", "No sabe"])]
df = df[~df["fami_trabajolaborpadre"].isin(["No aplica", "No sabe"])]
df = df[~df["fami_trabajolabormadre"].isin(["No aplica", "No sabe"])]
```

Con esto ahora contamos con 331342 filas.

Nos dimos a la tarea de reducir la dimensionalidad de la variable departamento, agrupándola según las regiones oficiales de Colombia. De esta forma, al aplicar posteriormente alguna técnica de codificación (encoding), se minimiza el impacto en la dimensionalidad que implicaría trabajar directamente con cada departamento. Adicionalmente, esta transformación contribuye a mitigar el desbalance en la distribución de valores.
